```python
# Refrence
# https://towardsdatascience.com/scraping-table-data-from-websites-using-a-single-line-in-python-ba898d54e2bc

import pandas as pd
import numpy as np
import time 
start = time.time()

dflink = pd.read_excel('link.xlsx')

a = "XX"
addrow1 = {
    'Agency' : [a],
    'Rating' : ['a'],
    'Outlook': ['b'],
    'Date' : ['c'],
    'Country' : ['d']
}

df1 = pd.DataFrame(addrow1)

i = 0
while i <152: 
    URL = dflink.Link[i]
    pd.read_html(URL)

    tables = pd.read_html(URL)
    #print(tables[0])
    print(URL)
    #print(len(tables))
    dftemp = tables[0]
    #print(dftemp)
    
    array = np.repeat(dflink.Country[i], len(tables[0]))
    #print(array)
    dftemp1 = dftemp.assign(Country = array)
    #print(dftemp1)

    df1 = df1.append(dftemp1)
    print(len(tables[0]), i)
    i = i +1


df1.drop(0)
df1.to_excel("fresh.xlsx")
end = time.time()
print(start)
print(end)
print(end-start)
df1
```
